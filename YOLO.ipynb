```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
img = 'https://thumbs.dreamstime.com/z/loving-tender-moment-men-horse-young-man-smiling-174963966.jpg'
```

```python
result = model(img)
```

```python
result.print()
```

```python
np.array(result.render()).shape
```

```python
plt.imshow(np.squeeze(result.render()))
plt.show()
```

```python
cap = cv2.VideoCapture(0)

while True:
    img , frame = cap.read()
    
    result = model(frame)
    
    cv2.imshow('Frame',np.squeeze(result.render()))
    
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid
import time
import os
```

```python
Images_path = os.path.join('data','images')

labels = ['Awake' , 'Drowsy']

num_imgs = 20
```

```python
#For collecting Images from webcam

cap = cv2.VideoCapture(0)

for label in labels:
    print("Collecting Images for {} :".format(label))
    time.sleep(5)
    
    for img_num in range(num_imgs):
        print("Collecting Images for {}, image number{} :".format(label,img_num))
        
        ret , frame = cap.read()
        
        imgname = os.path.join(Images_path, label+','+str(uuid.uuid1())+'.jpg')
        
        cv2.imwrite(imgname,frame)
        
        cv2.imshow('Image Collection', frame)
        time.sleep(2)
        
        k = cv2.waitKey(1)
        if k == ord('q'):
            break
        
cap.release()
cv2.destroyAllWindows()
```
